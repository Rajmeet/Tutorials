#### Data-Type (B): Sequences
A sequence is an ordered collection of items, indexed by positive integers.
1. STRING : This is an ordered sequence of letters enclosed by quotes and indexed by numbers
2. LIST : List is also a sequence of values of any type. Values in the list are called elements/items. 
3. TUPLE Tuples are a sequence of values of any type, and are indexed by integers. They are immutable. They are indicated by (). 

````python
a = "Hello"#String
l = [1,2,3] #List
t = (4,2) #Tuple
````

#### Data-Type (C): Sets
Sets is an unordered collection of values, of any time and cannot doesn't hold duplicate values. Syntax: set()


````python
l = [1,2,3,3] #multiple items removed 
s = set(l) 
print(s)
````

#### Data-Type (D): Mapping - Dictionary

### Flow of Statements (If/Else)
<b>If Condition</b>: The if condition, as the word says if something is true or false. Its syntax is: ```if(condition):``` <br>
<b>Else Condition</b>: If the if condition is not executed we go to the else condition. Its syntax is: ```else:``` <br>
<b>Elseif Condition</b>: If the if condition is executed and false then the program will run the elseif condition: Syntax ```elif(condition):```

> Why mutiple ifs are not recommended <br>
The elif command is command is more used instead of multiple if's in the program because in case of elseif the coded is executed and terminated as soon as a condition is met. However, if condition requires every block of if to be executed. Hence it is not efficient.
